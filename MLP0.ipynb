```python
chars = sorted(list(set(''.join(words))))
# Combine all unique characters, including special tokens, and map them to indices 0-27.
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
print(itos)
```

```python
block_size = 5
X, Y  = [], []

for w in words[:5]:
  print(w)
  context = [0] * block_size
  print(context)
  for ch in w + '.':
    ix = stoi[ch]
    X.append(context)
    Y.append(ix)
    print(''.join(itos[i] for i in context), '--->', itos[ix])
    context = context[1:] + [ix]

print(X)
print(Y)

```
